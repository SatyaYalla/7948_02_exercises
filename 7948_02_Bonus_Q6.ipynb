```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
imdb_df = pd.read_csv('C:/Users/satya.yalla/Documents/Tiger_Analytics_Learning_Path/Python_workspace/assignment/imdb.csv',escapechar="\\")
movie_df = pd.read_csv('C:/Users/satya.yalla/Documents/Tiger_Analytics_Learning_Path/Python_workspace/assignment/movie_metadata.csv')
```

```python
sns.heatmap(movie_df.corr(),cmap='coolwarm')
```

```python
sns.jointplot(x='year',y='imdbRating',data=imdb_df,kind='hex',cmap='twilight')
sns.jointplot(x='title_year',y='imdb_score',data=movie_df,kind='hex',cmap='twilight')
# most of the movies are rated in imdb in recent years
```

```python
sns.jointplot(x='title_year',y='gross',data=movie_df,kind='scatter',s=5)
sns.jointplot(x='imdb_score',y='gross',data=movie_df,kind='scatter',s=5)
# highest grossing movies in between 2000 and 2020
# movies with highest gross have high imdb ratings
```

```python
imdb_df = imdb_df.dropna()
movie_df = movie_df.dropna()
fig,axes = plt.subplots(1,2)
sns.kdeplot(imdb_df['imdbRating'],ax=axes[0])
sns.kdeplot(movie_df['imdb_score'],ax=axes[1])
#imdb_rating is kinda following normal distribution in both the dataframes
```

```python
fig,axes = plt.subplots(figsize=(20,4),nrows=1,ncols=4)
sns.barplot(x='type',y='nrOfWins',data=imdb_df,ax=axes[0])
sns.barplot(x='type',y='nrOfNominations',data=imdb_df,ax=axes[1])
sns.barplot(x='type',y='nrOfPhotos',data=imdb_df,ax=axes[2])
sns.barplot(x='type',y='nrOfNewsArticles',data=imdb_df,ax=axes[3])
#comparing the plots video.tv type has the highest no of wins, nominations and new articles
```

```python
sns.jointplot(x='duration',y='imdb_score',data=movie_df,s=9)
sns.jointplot(x='duration',y='title_year',data=movie_df,s=9)
# most videos have duration less than 150
```

```python
fig = plt.figure(figsize=(16,6))
sns.barplot(x='imdb_score',y='movie_facebook_likes',data=movie_df)
plt.tight_layout()
# more the number of FB likes more the imdb rating
```

```python
sns.violinplot(x='color',y='title_year',data=movie_df)
# color films started from around 1930 and are more frequent after 2000
# blakc&white films started way before 1920 and are more frequent during the 
```

```python
imdb_df['title_len']=imdb_df['title'].apply(lambda x:len(x))
sns.distplot(imdb_df['title_len'],kde=False,bins=30,color='r')
sns.jointplot(x='year',y='title_len',data=imdb_df,kind='kde',color='r')
# 20-30 is the most occuring movie length
```

```python
fig = plt.figure(figsize=(14,6))
sns.countplot(movie_df['country'])
fig.autofmt_xdate()
# Most of the movies in the dataframe are from USA
```
